```python
import pandas as pd
fpas = pd.read_excel('对比.xlsx', sheet_name = 0)
mses = pd.read_excel('对比.xlsx', sheet_name = 1)
nnzs = pd.read_excel('对比.xlsx', sheet_name = 2)
```

```python
df = pd.DataFrame()
```

```python
def func1(t):
    return round(t, 1)
def func4(t):
    return round(t, 4)
def func(t):
    return round(t, 0)
def fe(t):
    return "%.1e" % t
    
```

```python
df['filename'] = fpas['filename']
```

```python
df['nsga2-fpa'] = fpas['linear/nsga2/FPA_NNZ_MSE'].apply(func4)
df['nsga2-mse'] = mses['linear/nsga2/FPA_NNZ_MSE'].apply(fe)
df['nsga2-nnz'] = nnzs['linear/nsga2/FPA_NNZ_MSE'].apply(round)
```

```python
df['revised-nsga2-fpa'] = fpas['linear/nsga2_random20p_toZero/FPA_NNZ_MSE'].apply(func4)
df['revised_nsga2-mse'] = mses['linear/nsga2_random20p_toZero/FPA_NNZ_MSE'].apply(func1)
df['revised_nsga2-nnz'] = nnzs['linear/nsga2_random20p_toZero/FPA_NNZ_MSE'].apply(round)
```

```python
df['learning-to-rank-fpa'] = fpas['linear/CoDE'].apply(func4)
df['learning-to-rank-mse'] = mses['linear/CoDE'].apply(fe)
df['learning-to-rank-nnz'] = nnzs['linear/CoDE'].apply(round)
```

```python
df['lasso-fpa'] = fpas['lasso'].apply(func4)
df['lasso-mse'] = mses['lasso'].apply(func1)
df['lasso-nnz'] = nnzs['lasso'].apply(round)
```

```python
df['ridge-fpa'] = fpas['ridge'].apply(func4)
df['ridge-mse'] = mses['ridge'].apply(func1)
df['ridge-nnz'] = nnzs['ridge'].apply(round)
```

```python
df['lr-fpa'] = fpas['lr'].apply(func4)
df['lr-mse'] = mses['lr'].apply(func1)
df['lr-nnz'] = nnzs['lr'].apply(round)
```
