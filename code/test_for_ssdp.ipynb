```python
weights = []
bias = []
for i in range(input_size):
    weights.append(parameters[i*hidden_n:(i+1)*hidden_n])
for i in range(len(parameters)-hidden_n, len(parameters)):
    bias.append([parameters[i]])
```

```python
b = np.array(bias)
```
