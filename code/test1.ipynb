```python
doc1 = [['filename', 'FPA', 'numOfNonZero']] # 记录平均【FPA, 对应的非零参数解】
doc2 = []
doc4 = []
```

```python
print(params)
```

```python
params.to_csv('params.csv')
```

```python
avg_fpa = np.mean(model.NDSet.ObjV[:, 0]) # 记录平均【FPA, 对应的非零参数】
avg_nonp = np.mean(model.NDSet.ObjV[:, 1])
print('平均FPA, 对应非零参数个数', avg_fpa, avg_nonp)
doc1.append([file, avg_fpa, avg_nonp])

set_fpa = model.NDSet.ObjV[:, 0]
set_fpa = list(set_fpa)
set_fpa.insert(0, file)
print('非支配集的FPA', set_fpa)
doc2.append(set_fpa)

set_nonp = model.NDSet.ObjV[:, 1]
set_nonp = list(set_nonp)
set_nonp.insert(0, file)
print('非支配集的非零参数个数', set_nonp)
doc4.append(set_nonp)
```

```python
import BPNN
import helpers
from SSDP import SSDP
from BPNN import BPNN
file = 'ant-1.3'
path = 'data/'
X, y = helpers.getfeatures(path, file)
```

```python
model = BPNN(0.05, 500,X.shape[1], 5)
```

```python
print(X.shape, y.shape)
```

```python
model.fit(X, y)
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import target_functions as tgf
with tf.Session() as sess:
    sess.run(model.init)
    predValue = sess.run(model.final_opt, feed_dict = {model.X: X})
    print(predValue)
    fpa = tgf.FPA(predValue, y)
    print(fpa)
    aae = tgf.AAE(predValue, y)
    print(aae)
    y = y.reshape([-1,1])
    loss =sess.run(model.loss, feed_dict = {model.X:X, model.y:y})
    print(loss)
    
```
